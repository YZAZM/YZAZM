- 映射，将现有数据创建为新的表示数据，转为自己想要的格式。map()是返回新的，apply()是转换后的series.但不会改变原有数据结构

```python
reviews2_points_mean = reviews2.points.mean()
reviews2.points.map(lambda p: p-reviews2_points_mean)
```

```python
def remean_points(row):
    row.points = row.points - reviews2_points_mean
    return row
reviews2.apply(remean_points,axis = 'columns')
```

# Grouping and Sorting

## groupby()

- groupby()创建一组评论，该评论为给定得wine分配了相同得分数值，然后抓取points()并计算出现次数，value_counts()只是groupby()得快捷方法

```python
reviews2.groupby('points').points.count()
```

- 获取每个分值类别中最便宜的wine的价格

```python
reviews2.groupby('points').price.min()
```

- 从数据集中的每个酒庄中选择第一款评论的葡萄酒名称的一种方法

```python
reviews2.groupby('winery').apply(lambda df: df.title.iloc[0])
```

- 按国家和省份挑选组好的wine

```python
reviews2.groupby(['country','province']).apply(lambda df:df.loc[df.points.idxmax()])
```

### agg()方法

- 同时在dataframe上运行一堆不同的函数

```python
reviews2.groupby(['country']).price.agg([len,min,max])
```

## multi-indexes

- 多索引与常规索引的不同之处在于它具有多个级别

```python
countries_reviewed.reset_index()
```

```python
reviews2_written = reviews2.groupby(['taster_twitter_handle']).taster_twitter_handle.count()
reviews2_written
```

```python
best_rating_per_price = reviews2.groupby('price')['points'].max().sort_index()
best_rating_per_price
```

```python
best_rating_per_price = reviews2.groupby('price').points.max().sort_index()
best_rating_per_price
```

这里借助 groupby('price') 方法，按照 price（价格）这一属性对数据进行分组。其作用是把数据集中价格相同的记录归为一组。分组完成后，针对每个分组中的 points（评分）列运用  max()函数，这样就能得到每个价格组对应的最高评分。最后使用sort_index()方法，按照价格（也就是索引）对结果进行升序排列，从而让数据按照价格从低到高的顺序呈现。**同时注意上述两行代码的区别**

```python
price_extremes =reviews2.groupby(['variety']).price.agg([min,max])
price_extremes
```

```python
sorted_varieties = price_extremes.sort_values(by=['min','max'],ascending = False)
sorted_varieties
```

```python
reviewer_mean_ratings = reviews2.groupby(['taster_name']).points.mean()
reviewer_mean_ratings
```

```python
country_variety_counts = reviews2.groupby(['country','variety']).size().sort_values(ascending=False)
country_variety_counts
```

- size()函数，这样就能得到**每个分组中的记录数量**，也就是每个国家每种葡萄品种的出现次数。
